```python
import pandas as pd  
import numpy as np
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, LabelEncoder  
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
df = pd.read_csv(r"C:\Users\Mo\Desktop\3.3\assemant\assesment\WA_Fn-UseC_-Telco-Customer-Churn.csv")
df
```

```python
label_encode_cols = ["Partner", "Dependents", "PhoneService", "PaperlessBilling"]
one_hot_encode_cols = ["MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup",
                       "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies",
                       "Contract", "PaymentMethod"]
min_max_scale_cols = ["tenure", "MonthlyCharges", "TotalCharges"]
```

```python
X = df.drop(columns=['customerID', 'Churn'])
y = df['Churn']
```

```python
le_target = LabelEncoder()
y = le_target.fit_transform(y)
```

```python

joblib.dump(le_target, r"C:\Users\Mo\Desktop\3.3\label encoder target.pkl")
print(" saved.")
```

```python
X[min_max_scale_cols] = X[min_max_scale_cols].replace(' ', np.nan).astype(float)
X[min_max_scale_cols] = X[min_max_scale_cols].fillna(X[min_max_scale_cols].mean())
```

```python
label_encoders = {}
for col in label_encode_cols + ["gender"]:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
```

```python
joblib.dump(label_encoders, r"C:\Users\Mo\Desktop\3.3\label encoder.pkl")
print(" saved.")
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
one_hot_encoded = one_hot_encoder.fit_transform(X[one_hot_encode_cols])
```

```python
joblib.dump(one_hot_encoder, r"C:\Users\Mo\Desktop\3.3\one hot encoder.pkl")
print("saved.")
```

```python
min_max_scaler = MinMaxScaler()
scaled_numerical = min_max_scaler.fit_transform(X[min_max_scale_cols])
```

```python
joblib.dump(min_max_scaler, r"C:\Users\Mo\Desktop\3.3\minmax scaler.pkl")
print(" saved.")
```

```python
X_processed = np.hstack((X[label_encode_cols + ["gender"]].values, scaled_numerical, one_hot_encoded))

X_processed 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
model_file = r'C:\Users\Mo\Desktop\3.3\logistic_regression_model.pkl'
joblib.dump(model, model_file)
print(f"model saved as {model_file}.")
```
